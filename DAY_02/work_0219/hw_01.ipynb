1. 주어진 데이터의 공분산을 계산하세요


```python
def cov_mat(x, y):
    return round(np.cov(x, y, ddof=0)[0, 1], 3)
```

```python
for i in range(len(irisList)):
    for j in range(i, len(irisList)):
        if i != j:
            print(
                f"{irisChar[i]}와 {irisChar[j]}의 공분산 : {cov_mat(irisList[i], irisList[j])}"
            )
```

```python
irisDF.cov(numeric_only=True)
```

2. 주어진 데이터의 상관계수를 계산하세요


```python
def corr_mat(x, y):
    return round(np.corrcoef(x, y)[0, 1], 3)
```

```python
for i in range(len(irisList)):
    for j in range(i, len(irisList)):
        if i != j:
            print(
                f"{irisChar[i]}와 {irisChar[j]}의 공분산 : {corr_mat(irisList[i], irisList[j])}"
            )
```

```python
irisDF.corr(numeric_only=True)
```

3. 각 특성별로 히스토그램을 그려보세요


```python
def draw_hist(x: np.array, title: str, bins: int = 10):
    min_value = math.floor(x.min())
    max_value = math.ceil(x.max())
    ret = plt.hist(x, bins=bins, range=(min_value, max_value))
    plt.xticks(np.linspace(ret[1].min(), ret[1].max(), bins + 1))
    plt.title(title)
```

```python
plt.figure(figsize=(10, 8))
for i in range(4):
    plt.subplot(2, 2, i + 1)
    draw_hist(irisList[i], irisChar[i])
plt.tight_layout()
plt.show()
```

4. 각 품종별로 특성(꽃밪침 길이, 꽃받침 너비, 꽃잎 길이, 꽃잎 너비)에 대한 상자 그림(boxplot)을 그려보세요


```python
def draw_boxplot(species: str):
    data = []
    for feature in irisDF.columns[:4]:
        data.append(irisDF[irisDF["species"] == species][feature])
    plt.boxplot(data)
    plt.xticks([1, 2, 3, 4], irisDF.columns[:4])
    plt.title(species)
```

```python
plt.figure(figsize=(12, 6))
for idx, species in enumerate(irisDF["species"].unique(), 1):
    plt.subplot(1, 3, idx)
    draw_boxplot(species)
plt.tight_layout()
plt.show()
```

5. 각 품종별로 꽃받침 길이와 꽃받침 너비의 관계를 산점도로 나타내고, 색상으로 품종을 구분하세요


```python
plt.figure(figsize=(9, 8))
for species in irisDF["species"].unique():
    x = irisDF[irisDF["species"] == species]["sepal_length"]
    y = irisDF[irisDF["species"] == species]["sepal_width"]
    poly_fit = np.polyfit(x, y, deg=1)
    poly_1d = np.poly1d(poly_fit)
    xs = np.linspace(x.min(), x.max())
    ys = poly_1d(xs)
    plt.plot(xs, ys, label=f"{poly_fit[1]:.2f} + {poly_fit[0]:.2f}x", lw=0.5)
    plt.scatter(x, y, label=species)
    plt.legend()
plt.title("Sepal Length vs Sepal Width")
plt.show()
```
